###2.1 Let's begin by writing a function to compute the Root Mean Squared Error for a list of numbers


```python
"""
Function
--------
compute_rmse

Given two arrays, one of actual values and one of predicted values,
compute the Roote Mean Squared Error

Parameters
----------
predictions : array
    Array of numerical values corresponding to predictions
    for each of the N observations

yvalues : array
    Array of numerical values corresponding to the actual
    values for each of the N observations

Returns
-------
rmse : int
    Root Mean Squared Error of the prediction

Example
-------
>>> print compute_rmse((2,2,3),(0,2,6)
2.16
"""
def compute_rmse(predictions, yvalues):
    #print((predictions - yvalues)**2)
    rmse = np.sqrt(((predictions - yvalues) ** 2).mean())
    return float(rmse)

```

###3.1 Nearest Neighbors: Distance function
Let's try and build a machine learning algorithm to beat the "Average Value" baseline that we computed above.  Soon I will implement the Nearest Neighbor algorithm, but first we need to create a distance metric to measure the distance (and similarity) between two instances.

```python
"""
Function
--------
distance

Given two instances and a value for L, 
return the L-Norm distance between them

Parameters
----------
x1, x2 : array
    Array of numerical values corresponding to 
    predictions for each of the N observations

L: int
    Value of L to use in computing distances

Returns
-------
dist : int
    The L-norm distance between instances

Example
-------
>>> print distance((3,4),(6,8),2)
7

"""

def distance(x1, x2, L):
    # if arrays are not of equal length,
    # ignore the extra elements in larger array
    if len(x1) != len(x2):
        if len(x1) < len(x2):
            x2 = x2[:len(x1)]
        else:
            x1 = x1[:len(x2)]
    
    # Initialize sum variable with 0
    sum=0
    
    # Find the power sum of corresponding elements
    for i in range(0,len(x1)):
        sum += ((x1[i]-x2[i])**L)
    
    # Take L'th root of sum
    dist = (sum)**(1/float(L))
    
    return float(dist)
```

```python
# Example 1 - Euclidean distance in 2D
distance((3,4), (6,8), 2)
```

###3.2 Basic Nearest Neighbor algorithm

Let's start the fun now.  Here, I implement a basic nearest neighbor algorithm from scratch. Starting with a simple model that will use two input features (CRIM and RM) and a single output (MEDV).  In other words, I am modelling the relationship between median home value and crime rates and house size.

Use the training data (bdata_train) to "fit" the model, although as we know, with Nearest Neighbors there is no real training, we just need to keep your training data in memory.  Bellow is a function that predicts, for each instance in the testing data (bdata_test), the median home value using the nearest neighbor algorithm.  Since this is a small dataset, we can simply compare our test instance to every instance in the training set, and return the MEDV value of the closest training instance.  We also have the function take L as an input, where L is passed to the distance function.

Furthermore we compute the RMSE for the Nearest Neighbor predictions on the test set, using the Euclidean distance.

```python
"""
Function
--------
nneighbor

Given training and test data and a value for L,
return the predicted median home values

Parameters
----------
training data, test data : array
    Array of numerical values corresponding to (N-1)
    features and 1 result variable
    for both training and test datasets

L: int
    Value of L to use in computing distances

Returns
-------
medv_pred : array
    Array of predicted values for median home prices
rmase_nn : float
    Value for RMSE between observed and 
    predicted values for test data

Example
-------
>>> nneighbor(test_data,train_data,2)
[20.300000000000001,
 25.300000000000001,
 14.4,
 10.4,
 48.299999999999997,
 22.0,
 23.800000000000001,
 .....
 .....
 ]
 
6.618692249008643

"""

import time
def nneighbor(train_data, test_data, L):
    
    # Begin the timer
    start_time = time.time()
    
    # Declare an array to store median home 
    # value for all instances in test dataset
    medv_pred = []
    
    #########test_first = bdata_test.head(1)
    
    # Loop through all instances in test dataset
    for test_inst in test_data:
        
        # Declare a variable to store smallest distance,
        # thus initialize to infinity
        min_dist = np.Inf
           
        # Loop through entire training dataset for each of the test instances
        for train_inst in train_data:
            
            # Get Euclidean distance between test and training instance
            dist = distance(test_inst[:-1], train_inst[:-1], L)
            
            # Check if current distance is minimum between test instance
            # and training instance. If yes, store it as new minimum distance
            # and corresponding median home value as the prediction
            if dist < min_dist:
                min_dist = dist
                pred_value = train_inst[-1]
        
        # Append predicted median home value
        medv_pred.append(pred_value)
    
    # Stop the timer and print the time taken
    print("Time taken: " + str(round(time.time() - start_time,2)) + " seconds")
    
    # Compute RMSE for predicted and test values
    rmase_nn = compute_rmse(np.asarray(medv_pred),
                            test_data[:,(test_data.shape[1]-1)])
    
    return medv_pred, rmase_nn
```

```python
# Without normalization
train_data = bdata_train[['CRIM','RM','TARGET']].values
test_data = bdata_test[['CRIM','RM','TARGET']].values

medv_pred, rmase_nn = nneighbor(train_data,test_data,2)
print('Root Mean Square Error for the basic KNN model:', rmase_nn)
# Plot the observed and predicted values on a scatter plot
plot_obs_pred(test_data[:,(test_data.shape[1]-1)], np.asarray(medv_pred))
```

###3.3 Results and Normalization

We did not really normalized our features - a big no-no with Nearest Neighbor algorithms.  Let's write a generic normalization function that takes as input an array of values for a given feature, and returns the normalized array (subtract the mean and divide by the standard deviation).

Then we re-run the Nearest Neighbor algorithm on the normalized dataset (still just using CRIM and RM as input), and compare the RMSE from this method with your previous RMSE evaluations.

```python
# Function to normalize each column of input array
# using the column's mean and standard deviation
"""
Function
--------
normalize

 Takes as input an array of values for a given feature,
 and returns the normalized array 
 (subtract the mean and divide by the standard deviation).

Parameters
----------
data : array
    Array of numerical values

Returns
-------
data : array
    normalized array

Example
-------
>>> normalize(bdata_train[['ZN','AGE','TARGET']].values)
array([[ -3.81161053e-01,  -8.13382747e-01,   1.87000000e+01],
       [ -3.80561354e-01,  -2.43503114e-01,   2.33000000e+01],
       [ -2.66918872e-01,  -1.84170516e+00,   1.53000000e+01],
       .....
       .....
       ]

"""

def normalize(data):
    # For each column, normalize each value
    for k in range(0,data.shape[1]-1):
        data[:,k] = (data[:,k]-data[:,k].mean())/(data[:,k].std())
    return data
```

```python
# With normalization
train_data = normalize(bdata_train[['CRIM','RM','TARGET']].values)
test_data = normalize(bdata_test[['CRIM','RM','TARGET']].values)

medv_pred, rmase_nn = nneighbor(train_data,test_data,2)
print('Root Mean Square Error for the basic KNN model:', rmase_nn)
# Plot the observed and predicted values on a scatter plot
plot_obs_pred(test_data[:,(test_data.shape[1]-1)], np.asarray(medv_pred))
```

### 3.5 Cross-Validation

The more we tinkered with your features and distance function, the higher the risk that you overfit your training data. To prevent this sort of overfitting, we need to use cross-validation.

Let's use 10-fold cross-validation and see the average RMSE for Nearest Neighbors using Euclidean distance with CRIM and RM input features, as well as the total running time for the full run of 10 folds.

```python
"""
Function
--------
cross_validation

 Takes as input a dataset, a value for L, a value for K,
 type of algorhtm (basicKNN or KNN), return the 
 RMSE value and runtime for execution

Parameters
----------
data : array
    Array of numerical values

Returns
-------
RMSE:
    RMSE value
runtime:
    Time taken for execution

Example
-------
>>> cross_validation(normalize(bdata_df[['CRIM','RM','TARGET']].values),2,3,"basic_knn")
Total Time taken for cross-validation:  0.85  seconds
Mean RMSE:  7.458960636085439

"""
def cross_validation(data,L,K,algorithm):
    
    np.random.seed(seed=13579)
    np.random.shuffle(data)
    total_rmse = 0
    
    # Begin the timer
    start_time = time.time()
    
    # Loop through 10 times, selecting different test (1 part) 
    # and training (9 parts) datasets for each iteration
    for i in range(10):
        test_range = range(int((i/10.0)*data.shape[0]),
                           int(((i+1)/10.0)*data.shape[0]))
        test = data[test_range,:]
        train = np.delete(data,test_range,0)
        if algorithm.lower() == "basic_knn":
            medv_pred, rmse = nneighbor(train,test,L)
        if algorithm.lower() == "knn":
            medv_pred, rmse, runtime = knn(train,test,L,K)
            #print("Not implemented yet")
        
        total_rmse += rmse
    
    # Stop the timer and calculate runtime
    runtime = str(round(time.time() - start_time,2))
    
    # Get the avarage of RMSE total
    mean_rmse = total_rmse/10
    
    return mean_rmse, runtime
```

```python
# Use 10-fold cross validation for CRIM and RM
# features using the basic Nearest Neighbors model.
mean_rmse, runtime = cross_validation(normalize(bdata_df[['CRIM','RM','TARGET']].values),
                                      2,3,"basic_knn")
# Stop the timer and print the time taken
print("Total Time taken for cross-validation: ",runtime," seconds")
print("Mean RMSE: ", mean_rmse)
```

Total time taken for above process is 0.95 seconds, which is more than the time taken without using cross-validation, since we are repeating the process 10 times.

###3.6 K-Nearest Neighbors Algorithm

Finally, let us implement the K-Nearest Neighbors algorithm.  Using 10-fold cross validation, let's see the RMSE for K=3 and the running time of the algorithm.

```python
"""
Function
--------
knn

 Takes as input training and test datasets,
 a value for L, a value for K,
 return the predicted median home values, RMSE and time taken.

Parameters
----------
train_data : array
    Array of numerical values
test_data : array
    Array of numerical values

Returns
-------
med_pred : array
    predictions of median home values
rmse_knn:
    RMSE value
runtime:
    time taken for execution

Example
-------
>>> knn(train,test,L,K)

"""

def knn(train_data, test_data, L, K):
    
    # Begin the timer
    stime = time.time()
    medv_pred = []
    
    # Loop for each instance in test data
    for test_inst in test_data:
            neighbors = []
            # Loop for each instance in training data
            for train_inst in train_data:
                # Get Eucledian distance
                dist = distance(train_inst[:-1],test_inst[:-1],L)
                # Append median home value and distance to list
                neighbors.append((train_inst[train_inst.shape[0]-1], dist))
            # Sort based on distance in ascending order
            neighbors.sort(key = operator.itemgetter(1),reverse = False)
            # Take average of median house values based on K
            medv_pred.append(np.mean(neighbors[0:K],axis = 0))
    medv_pred = np.array(medv_pred)
    runtime = np.round(time.time()-stime,2)
    
    # Get RMSE
    rmse_knn = compute_rmse(test_data[:,(test_data.shape[1]-1)],medv_pred[:,0])
    
    return medv_pred, rmse_knn, runtime
```

```python
# Use KNN with K=3 and cross validation for predicting median home values using CRIM and RM featues.

mean_rmse, runtime = cross_validation(normalize(bdata_df[['CRIM','RM','TARGET']].values),2,3,"knn")
print("Time taken:", runtime, "seconds")
print("RMSE using 10-fold cross-validation for K=3:", mean_rmse)
```

###3.7 Using cross validation to find K

What is the best choice of K?  Let's compute the RMSE for values of K between 1 and 25 using 10-fold cross-validation. Let us use the following features in your model: CRIM, ZN, RM, AGE, DIS, TAX.

```python
# Declare a list to store RMSE values and corresponding K value
rmse_k = []
# K starts from 1
k = 1

# For each value of K, perform cross-validation 
# on 6 features to predict median house values
# using KNN and L = 2 
for i in range(1,26):
    mean_rmse, runtime = cross_validation(normalize(
        bdata_df[['CRIM','ZN','RM','AGE','DIS','TAX','TARGET']].values)
                                          ,2,i,'knn')
    rmse_k.append((k,mean_rmse))
    k += 1   
# Make an array from the list of tuples
rmse_k = np.asarray(rmse_k)

# Plot RMSE values and corresponding K values
plt.xticks(np.arange(min(rmse_k[:,0]), max(rmse_k[:,0])+1, 1.0))
plt.plot(rmse_k[:,0],rmse_k[:,1])
plt.xlabel("K Values")
plt.ylabel("RMSE Values")
plt.title("RMSE vs K values for L=2")
plt.show()
```
